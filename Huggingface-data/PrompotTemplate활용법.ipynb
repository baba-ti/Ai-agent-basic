```python
from langchain_upstage import ChatUpstage

llm = ChatUpstage(model="solar-pro3", temperature=0)

question = "인프런에는 어떤 강의가 있나요?"
# llm.invoke(query)
```

```python
text_prompt = f"아래 질문에 답변해주세요: {question}"

llm.invoke(text_prompt)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_with_template = "아래 질문에 답변해주세요:\n\n {question}"

prompt_template = PromptTemplate(template=text_prompt, input_variables=["question"])
prompt_template.invoke({"question": question})
```

```python
llm.invoke(prompt_template.invoke({"question": question}))
```

```python
prompt_chain = prompt_template | llm
prompt_chain.invoke({"question": question})
```

```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.inflearn.com/courses/it-programming")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1500, chunk_overlap=200)

document_list = loader.load_and_split(text_splitter)
```

```python
from langchain_chroma import Chroma
from langchain_upstage import UpstageEmbeddings

embedding = UpstageEmbeddings(model='solar-embedding-1-large')
database = Chroma.from_documents(
    documents=document_list, 
    embedding=embedding,
    collection_name="inflearn_lectures",
    persist_directory="./inflearn-lectures"
    )
```

```python

retriever = database.as_retriever(search_kwargs={"k": 1})
retriever.invoke(question)
retriever_docs = retriever.invoke(question)
retriever_docs
```

```python
rag_prompt_template = PromptTemplate(
    template="""
    You are a helpful assistant that can answer questions about the Inflearn website.
    You are given the follwoing context:
    {context}
    Question: {question}
    """,
    input_variables=["context", "question"]
)
rag_chain = rag_prompt_template | llm

rag_chain.invoke({
    "context": retriever_docs,
    "question": question
})
```

```python
from langchain_classic import hub

rlm_rag_prompt = hub.pull("rlm/rag-prompt")
```

```python
rlm_rag_chain = rlm_rag_prompt | llm

rlm_rag_chain.invoke({
    "context": retriever_docs,
    "question": question
})
```

```python
from langchain_core.runnables import RunnablePassthrough

rag_chain =(
    {"context": retriever, "question": RunnablePassthrough()}|
    rlm_rag_prompt | llm
)

rag_chain.invoke(question)
```
